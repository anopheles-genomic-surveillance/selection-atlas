# Alert SA-AFUN-04 (*Rdl*)

This alert reports selection signals on Chromosome 3RL within the region 13,537,662-13,737,363 bp.

## Selection signals

{term}`Selection signal`s overlapping this {term}`genome region` are shown in the figure below.

```python
df_signals = page_utils.load_signals(
    contig=region_contig,
    start=region_start,
    stop=region_stop,
)

gene_labels = dict()
for item in alert["ir_candidate_genes"]:
    g = item["identifier"]
    gene_labels[g] = " "

if len(df_signals) > 0:
    page_utils.plot_signals(
        df=df_signals,
        contig=region_contig,
        x_min=df_signals["span2_pstart"].min() - 50_000,
        x_max=df_signals["span2_pstop"].max() + 50_000,
        gene_labels=gene_labels,
        genes_height=90,
    )
else:
    display(Markdown("No signals found."))
```

## Insecticide resistance genes

The following {term}`gene`s are found within this {term}`genome region` and may be driving 
{term}`recent positive selection` based on evidence for an association with 
{term}`insecticide resistance`. Please note that other genes are also within the affected 
genome region and may be driving selection.
### <a href='https://vectorbase.org/vectorbase/app/record/gene/LOC125769835' target='_blank'>LOC125769835</a> (*Rdl*, *AFUN2_014608*, *AFUN016339*)

This gene encodes a subunit of a GABA-gated chloride channel which is the binding target of cyclodiene insecticides including dieldrin and lindane. Mutations in this gene are associated with dieldrin resistance in multiple insect species, including A296G and A296S in *Anopheles gambiae* (codon 302 in *Drosophila melanogaster*). Resistance alleles persist in *Anopheles* populations in sub-Saharan Africa despite a ban on dieldrin use in the 1980s.
 See also:
